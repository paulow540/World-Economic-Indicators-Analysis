```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import plotly.express as px
```

```python
wd = pd.read_excel("C:\\Users\\Administrator\\Desktop\\world econonmic indicators dataset\\WorldBank.xlsx")
wd.head(4)
```

```python
hd = pd.read_csv("C:\\Users\\Administrator\\Desktop\\world econonmic indicators dataset\\HDI.csv")
hd.head(3)
```

```python
hd.columns[:4]
```

```python
hdi = pd.melt(hd, id_vars=hd.columns[:4], var_name='Hdi year', value_name='Index')
```

```python
hdi.head()
```

```python
hdi["hdicode"] = hdi["hdicode"].fillna("Blank")
hdi["region"] = hdi["region"].fillna("Blank")
hdi["Index"] = hdi["Index"].fillna(0)
```

```python
hdi.isna().sum()
```

```python
wd.info()
```

```python
for i in wd.dtypes[wd.dtypes == 'float64'].index.tolist():
    wd[i] = wd[i].fillna(0)
```

```python
hdi.info()
```

```python
# merge
final_df = pd.merge(hdi,wd, left_on="iso3", right_on="Country Code")
final_df
```

```python
final_df["country"].nunique()
```

# EDA

```python
country_birth_rate = final_df.groupby("country")["Birth rate, crude (per 1,000 people)"].sum().sort_values(ascending=False)[:10]
fig = px.bar(country_birth_rate, title="Birth rate, crude (per 1,000 people) by country")
fig.show(renderer="iframe")
```

```python
country_death_rate = final_df.groupby("country")["Death rate, crude (per 1,000 people)"].sum().sort_values(ascending=False)[:10]
fig = px.bar(country_death_rate, title="Death rate, crude (per 1,000 people) by country")
fig.show(renderer="iframe")
```

```python
country_unemploy = final_df.groupby("country")["Unemployment (% of total labor force) (modeled ILO estimate)"].sum().sort_values(ascending=False)[:10]
fig = px.bar(country_unemploy, title="Unemployment (% of total labor force) (modeled ILO estimate) by country")
fig.show(renderer="iframe")
```

```python
import pandas as pd
import plotly.graph_objects as go


years = sorted(final_df['Year'].unique())
fig = go.Figure()

# Create a trace for each year
for i, year in enumerate(years):
    top_artists = final_df[final_df["Year"] == year].groupby("country")["Unemployment (% of total labor force) (modeled ILO estimate)"].sum().sort_values(ascending=False)[:10]

    fig.add_trace(go.Bar(
        x=top_artists.index,
        y=top_artists.values,
        name=str(year),
        visible=(i == 0)  # Show only the first year initially
    ))

# Create buttons for dropdown
buttons = [
    dict(label=str(year),
         method="update",
         args=[{"visible": [j == i for j in range(len(years))]},
               {"title": f"Top Artists in {year}"}])
    for i, year in enumerate(years)
]

# Update layout with dropdown
fig.update_layout(
    updatemenus=[dict(
        active=0,
        buttons=buttons,
        x=0.1,
        y=1.2
    )],
    title="Top Artists by Year",
    xaxis_title="Artist Name",
    yaxis_title="Count",
    showlegend=False
)

fig.show(renderer='iframe')
```
